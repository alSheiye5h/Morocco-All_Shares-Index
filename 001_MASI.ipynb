```python
import numpy as np
import pandas as pd
import math
import CasaBourseScrap  as cbs
from CasaBourseScrap.quick import scrape_casablanca_bourse_js_click
from CasaBourseScrap.Notation import name_from_ticker, ticker_from_name
from datetime import datetime, timedelta
```

```python
stocks_brute = scrape_casablanca_bourse_js_click()
```

```python
stocks = stocks_brute.drop(['Code_ISIN', 'Emetteur', 'Instrument', 'Categorie'], axis=1)
stocks
```

```python

end_date = datetime.today().date()
start_date = end_date - timedelta(days=1) 
 
names = stocks['Ticker'].apply(lambda x: name_from_ticker(x))

full_data, latest_prices = cbs.loadmany(
    list(names), 
    start=start_date.isoformat(), 
    end=end_date.isoformat(), 
    method="cloudscraper"
)
```

```python
stocks_all = stocks.merge(
    latest_prices.rename('close').reset_index(),
    left_on='Ticker',
    right_on='ticker',
    how='left'
).drop('ticker', axis=1)

stock_nan = stocks_all[stocks_all['close'].isna()]
stock_filtred = stocks_all.dropna(subset='close')
```
