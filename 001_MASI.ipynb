```python
import numpy as np
import pandas as pd
import math
import CasaBourseScrap  as cbs
from CasaBourseScrap.quick import scrape_casablanca_bourse_js_click
from CasaBourseScrap.Notation import name_from_ticker
from datetime import datetime, timedelta
```

```python
stocks_brute = scrape_casablanca_bourse_js_click()
```

```python
stocks = stocks_brute.drop(['Code_ISIN', 'Emetteur', 'Instrument', 'Categorie'], axis=1)
stocks
```

```python

end_date = datetime.today().date()
start_date = end_date - timedelta(days=1) 
 
names = stocks['Ticker'].apply(lambda x: name_from_ticker(x))

data = cbs.loadmany(list(names), start=start_date.isoformat(), end=end_date.isoformat(), method="cloudscraper")
data
```

```python


stocks = stocks.merge(
    data[['ticker', 'value']].rename(columns={'value': 'close'}),
    left_on='Ticker',
    right_on='ticker',
    how='left'
)

stocks



# try:
#     # Get only the most recent day (latest price)
#     end_date = datetime.today().date()
#     start_date = end_date - timedelta(days=1)  # Just get yesterday and today
    
#     print(f"Fetching latest price data: {start_date} to {end_date}")
    
#     data = cbs.load_data('Addoha', start=start_date.isoformat(), end=end_date.isoformat(), method="cloudscraper")
    
#     if data is not None and hasattr(data, 'shape') and data.shape[0] > 0:
#         print("‚úÖ Success! Data shape:", data.shape)
        
#     else:
#         print("‚ùå No data retrieved or empty dataset")
        
# except Exception as e:
#     print(f"Error: {e}")

# data
```
