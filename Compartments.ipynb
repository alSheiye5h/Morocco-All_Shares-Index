# Comprehensive Compartment Analysis Formulas


## 1. **Risk-Adjusted Return Matrix**
- Principal A vs Principal B = (Avg Return_A ÷ Risk_A) ÷ (Avg Return_B ÷ Risk_B)
- Principal A vs Principal F = (Avg Return_A ÷ Risk_A) ÷ (Avg Return_F ÷ Risk_F)
- Principal A vs Alternatif A = (Avg Return_A ÷ Risk_A) ÷ (Avg Return_AltA ÷ Risk_AltA)
- Principal B vs Principal F = (Avg Return_B ÷ Risk_B) ÷ (Avg Return_F ÷ Risk_F)
- Principal B vs Alternatif A = (Avg Return_B ÷ Risk_B) ÷ (Avg Return_AltA ÷ Risk_AltA)
- Principal F vs Alternatif A = (Avg Return_F ÷ Risk_F) ÷ (Avg Return_AltA ÷ Risk_AltA)

## 2. **Market Efficiency Score**
- Principal A Score = (Liquidity_A × 0.3) + (Market Cap Growth_A × 0.4) + (Dividend Yield_A × 0.3)
- Principal B Score = (Liquidity_B × 0.3) + (Market Cap Growth_B × 0.4) + (Dividend Yield_B × 0.3)
- Principal F Score = (Liquidity_F × 0.3) + (Market Cap Growth_F × 0.4) + (Dividend Yield_F × 0.3)
- Alternatif A Score = (Liquidity_AltA × 0.3) + (Market Cap Growth_AltA × 0.4) + (Dividend Yield_AltA × 0.3)

## 3. **Performance Multiples Matrix**
- Principal A vs Principal B = Total Return_A ÷ Total Return_B
- Principal A vs Principal F = Total Return_A ÷ Total Return_F
- Principal A vs Alternatif A = Total Return_A ÷ Total Return_AltA
- Principal B vs Principal F = Total Return_B ÷ Total Return_F
- Principal B vs Alternatif A = Total Return_B ÷ Total Return_AltA
- Principal F vs Alternatif A = Total Return_F ÷ Total Return_AltA

## 4. **Risk-Reward Profile**
- Principal A Quality = Average Annual Return_A ÷ Price Volatility_A
- Principal B Quality = Average Annual Return_B ÷ Price Volatility_B
- Principal F Quality = Average Annual Return_F ÷ Price Volatility_F
- Alternatif A Quality = Average Annual Return_AltA ÷ Price Volatility_AltA

## 5. **Liquidity Premium Matrix**
- Principal A vs Principal B = Daily Volume_A ÷ Daily Volume_B
- Principal A vs Principal F = Daily Volume_A ÷ Daily Volume_F
- Principal A vs Alternatif A = Daily Volume_A ÷ Daily Volume_AltA
- Principal B vs Principal F = Daily Volume_B ÷ Daily Volume_F
- Principal B vs Alternatif A = Daily Volume_B ÷ Daily Volume_AltA
- Principal F vs Alternatif A = Daily Volume_F ÷ Daily Volume_AltA

## 6. **Growth Potential Index**
- Principal A Growth = (Revenue Growth_A + Earnings Growth_A + Market Cap Growth_A) ÷ 3
- Principal B Growth = (Revenue Growth_B + Earnings Growth_B + Market Cap Growth_B) ÷ 3
- Principal F Growth = (Revenue Growth_F + Earnings Growth_F + Market Cap Growth_F) ÷ 3
- Alternatif A Growth = (Revenue Growth_AltA + Earnings Growth_AltA + Market Cap Growth_AltA) ÷ 3

## 7. **Stability Metric**
- Principal A Stability = 1 ÷ Price Volatility_A
- Principal B Stability = 1 ÷ Price Volatility_B
- Principal F Stability = 1 ÷ Price Volatility_F
- Alternatif A Stability = 1 ÷ Price Volatility_AltA

## 8. **Value Comparison Matrix**
- Principal A vs Principal B = Market Cap_A ÷ Market Cap_B
- Principal A vs Principal F = Market Cap_A ÷ Market Cap_F
- Principal A vs Alternatif A = Market Cap_A ÷ Market Cap_AltA
- Principal B vs Principal F = Market Cap_B ÷ Market Cap_F
- Principal B vs Alternatif A = Market Cap_B ÷ Market Cap_AltA
- Principal F vs Alternatif A = Market Cap_F ÷ Market Cap_AltA

## 9. **Income Generation Ratio**
- Principal A vs Principal B = Total Dividend Income_A ÷ Total Dividend Income_B
- Principal A vs Principal F = Total Dividend Income_A ÷ Total Coupon Income_F
- Principal A vs Alternatif A = Total Dividend Income_A ÷ Total Dividend Income_AltA
- Principal B vs Principal F = Total Dividend Income_B ÷ Total Coupon Income_F
- Principal B vs Alternatif A = Total Dividend Income_B ÷ Total Dividend Income_AltA
- Principal F vs Alternatif A = Total Coupon Income_F ÷ Total Dividend Income_AltA

## 10. **Composite Quality Score**
- Principal A Composite = (Risk-Reward_A × 0.25) + (Growth Score_A × 0.25) + (Stability_A × 0.25) + (Liquidity Score_A × 0.25)
- Principal B Composite = (Risk-Reward_B × 0.25) + (Growth Score_B × 0.25) + (Stability_B × 0.25) + (Liquidity Score_B × 0.25)
- Principal F Composite = (Risk-Reward_F × 0.25) + (Growth Score_F × 0.25) + (Stability_F × 0.25) + (Liquidity Score_F × 0.25)
- Alternatif A Composite = (Risk-Reward_AltA × 0.25) + (Growth Score_AltA × 0.25) + (Stability_AltA × 0.25) + (Liquidity Score_AltA × 0.25)

## **Interpretation Guide:**
- **Ratio > 1**: First compartment is "better" than second
- **Ratio = 1**: Compartments are equal
- **Ratio < 1**: Second compartment is "better" than first
- **Higher Scores**: Better overall performance
