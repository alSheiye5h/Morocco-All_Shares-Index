```python
import CasaBourseScrap  as cbs
from datetime import datetime, timedelta
```

```python
cbs.notation()
```

```python
try:
    # Get last 7 days
    end_date = datetime.today().date()
    start_date = end_date - timedelta(days=7)
    
    print(f"Fetching last week data: {start_date} to {end_date}")
    
    data = cbs.load_data('BCP', start=start_date.isoformat(), end=end_date.isoformat(), method="cloudscraper")
    print("Success! Data shape:", data.shape if hasattr(data, 'shape') else "No shape info")
    print(data.head() if data is not None else "No data")
    
except Exception as e:
    print(f"Error: {e}")
```
